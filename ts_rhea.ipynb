```python
ts1_raw = pd.read_csv("final_timeseries_1.csv")
ts1_raw.head()
```

```python
ts1 = pd.read_csv(
    "final_timeseries_1.csv",
    parse_dates=["date"]      # we know the column name is 'date'
)

# set date as index
ts1 = ts1.set_index("date")

# sort by date just to be safe
ts1 = ts1.sort_index()

print("ts1 index dtype:", ts1.index.dtype)
ts1.head()
```

```python
ts2 = pd.read_csv("final_timeseries_2.csv")

# If there is an extra unnamed index column, drop it
if "Unnamed: 0" in ts2.columns:
    ts2 = ts2.drop(columns=["Unnamed: 0"])

# Convert the 'date' column to datetime
ts2["date"] = pd.to_datetime(ts2["date"])

# Set 'date' as the index and sort
ts2 = ts2.set_index("date")
ts2 = ts2.sort_index()

print("ts2 columns:", ts2.columns)
print("ts2 index dtype:", ts2.index.dtype)
ts2.head()
```

```python
min_date = ts1.index.min()
max_date = ts1.index.max()

full_range = pd.date_range(start=min_date, end=max_date, freq="D")

min_date, max_date, len(full_range)
```

```python
missing_dates_ts1 = full_range.difference(ts1.index)
missing_dates_ts2 = full_range.difference(ts2.index)

print("Missing dates in ts1:", len(missing_dates_ts1))
print("Missing dates in ts2:", len(missing_dates_ts2))
```

```python
ts1_full = ts1.reindex(full_range)
ts2_full = ts2.reindex(full_range)

na_after_ts1 = ts1_full.isna().sum().sum()
na_after_ts2 = ts2_full.isna().sum().sum()

print("Missing values in ts1 after reindex:", na_after_ts1)
print("Missing values in ts2 after reindex:", na_after_ts2)
```

```python
na_before_ts1 = ts1.isna().sum().sum()
na_before_ts2 = ts2.isna().sum().sum()

print("Missing values in ts1 before reindex:", na_before_ts1)
print("Missing values in ts2 before reindex:", na_before_ts2)
```
